```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn import metrics 
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from scipy.stats import randint as sp_randint
from scipy.stats import uniform as sp_uniform
from sklearn.neural_network import MLPClassifier
import warnings
warnings.filterwarnings("ignore")

from sklearn.metrics import mean_squared_error, r2_score

import eli5
from eli5.sklearn import PermutationImportance
```

```python
df = pd.read_csv('Model_data.csv')
```

```python
df.head()
```

```python
pd.set_option('display.max_columns', 500)
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df.head()
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split


train, test = train_test_split(df, test_size=0.25, random_state=25)
```

```python
x_train = train[['Age', 'Overall', 'centimeter height', 'Weight',
       'International Reputation', 'Weak Foot',
       'Skill Moves', 'Crossing', 'Finishing', 'HeadingAccuracy',
       'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility',
       'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength',
       'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Penalties', 'Marking', 'StandingTackle', 'SlidingTackle']]
y_train = train[['Peak Overall']]

x_test = test[['Age', 'Overall', 'centimeter height', 'Weight',
       'International Reputation', 'Weak Foot',
       'Skill Moves', 'Crossing', 'Finishing', 'HeadingAccuracy',
       'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility',
       'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength',
       'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Penalties', 'Marking', 'StandingTackle', 'SlidingTackle']]
y_test = test[['Peak Overall']]
```

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()  
scaler.fit(x_train)  
x_train_mlp = scaler.transform(x_train)  
x_test_mlp = scaler.transform(x_test)  

scaler = StandardScaler()
scaler.fit(y_train)
y_train_mlp = scaler.transform(y_train)
y_test_mlp = scaler.transform(y_test)
y_expert = test[['Expert Potential']]
y_expert_mlp = scaler.transform(y_expert)
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
```

```python
LR = LinearRegression()
lr = LR.fit(x_train_mlp, y_train_mlp)
```

```python
y_pred = lr.predict(x_test_mlp)
```

```python
m1 = mean_squared_error(y_test_mlp, y_pred)
m2 = mean_squared_error(y_test_mlp, y_expert_mlp)
print('Linear Regression Model\'s MSE is', m1)
print('Expert Guess\'s MSE is', m2)
```

```python
perm = PermutationImportance(lr, random_state=25).fit(x_test_mlp, y_test_mlp)
eli5.show_weights(perm, feature_names = x_test.columns.tolist(), top=35)
```

```python
SGD = SGDRegressor(random_state=25)
param_dist1 = {'penalty': ['l1', 'l2'], 'alpha': sp_uniform(1e-5, 10.0)}
sgd_lr = RandomizedSearchCV(SGD, param_dist1, 
            n_iter=200, n_jobs=-1, cv=5, random_state=25, scoring='neg_mean_squared_error')
sgd = sgd_lr.fit(x_train_mlp, y_train_mlp)
```

```python
print('Optimal tuning parameter values:\n', sgd.best_params_)
print('MSE of the optimal results:', abs(sgd.best_score_))
```

```python
y_pred = sgd.predict(x_test_mlp)
m1 = mean_squared_error(y_test_mlp, y_pred)
m2 = mean_squared_error(y_test_mlp, y_expert_mlp)
print('Linear Regression Model\'s MSE is', m1)
print('Expert Guess\'s MSE is', m2)
```

```python
perm = PermutationImportance(sgd, random_state=25).fit(x_test_mlp, y_test_mlp)
eli5.show_weights(perm, feature_names = x_test.columns.tolist(), top=35)
```

```python
param_dist2 = {'n_estimators': sp_randint(10, 200),
               'max_depth': sp_randint(2, 4),
               'min_samples_split': sp_randint(2, 20),
               'min_samples_leaf': sp_randint(2, 20),
               'max_features': sp_randint(1, 4)}
RFC = RandomForestRegressor(bootstrap=True,oob_score=True, random_state=25)
rscv_rf = RandomizedSearchCV(RFC, param_dist2, 
         n_iter=200, n_jobs=-1, cv=5, random_state=25, scoring='neg_mean_squared_error')

rf = rscv_rf.fit(x_train_mlp, y_train_mlp)
```

```python
print('Optimal tuning parameter values:\n', rf.best_params_)
print('MSE of the optimal results:', abs(rf.best_score_))
```

```python
y_pred = rscv_rf.predict(x_test_mlp)


m1 = mean_squared_error(y_test_mlp, y_pred)

m2 = mean_squared_error(y_test_mlp, y_expert_mlp)
print('Random Forest Regression Model\'s MSE is', m1)
print('Expert Guess\'s MSE is', m2)
```

```python
perm = PermutationImportance(rf, random_state=25).fit(x_test_mlp, y_test_mlp)
eli5.show_weights(perm, feature_names = x_test.columns.tolist(), top=35)
```

```python
param_dist3 = {'C': sp_uniform(loc=0.1, scale=10.0), 'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
               'gamma': ['scale', 'auto'],
               'shrinking': [True, False]}
svr = SVR()
rscv_SVR = RandomizedSearchCV(svr, param_dist3, 
        n_iter=200, n_jobs=-1, cv=5, random_state=25, scoring='neg_mean_squared_error')

random_SVR = rscv_SVR.fit(x_train_mlp, y_train_mlp)
```

```python
perm = PermutationImportance(random_SVR, random_state=25).fit(x_test_mlp, y_test_mlp)
eli5.show_weights(perm, feature_names = x_test.columns.tolist(), top=35)
```
